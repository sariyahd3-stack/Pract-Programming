# **Concatenation**

```python
FULLNAME="Sariyah" + "Dublin" #Concatentate 2 strings 
print(FULLNAME)

#FULLNAME="Dublin"+"Sariyah" #concatenate 2 strings in reverse 
#print(FULLNAME)
```

```python
REVERSE='' #define empty string

for CHAR in FULLNAME: #for characters in string FULLNAME
    REVERSE = CHAR + REVERSE #concatinate the characters of the string to 

print(f"My name spelled backwards is: {REVERSE}")
```

When I concatenate, the order of the strings matter.

So, if I loop through a string, concatenate the first character of the string to a new string, and then keep looking through the character, concatenating to the beginning of the string, it will reverse the string
