Slicing is splitting a string into charachters. You can index into a string just liek you can index into a list (or a nested list). We can slice one character, or many characters)

```python
print(FIRST)
print('The LAST letter of my name is:', FIRST[6])
```

```python
#slicing many characters from a string at once 
FULLNAME='Sariyah Dublin'
print(FULLNAME)
NEW_FIRST=FULLNAME[0:7] #will show characters 0-6, last number is making sure not including whatever is in that index
NEW_LAST=FULLNAME[8:14] #characters 8-13
print('First name:', NEW_FIRST)
print('Last name:', NEW_LAST)
```
